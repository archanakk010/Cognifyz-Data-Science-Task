#  Level 3 - Task 2: Customer Preference Analysis

# Overview of Problem Statement

Customer preferences vary by cuisine, and understanding these patterns can help restaurants and platforms tailor their offerings for better engagement and satisfaction.

# Objective

1.Analyze the relationship between cuisine type and ratings

2.Identify the most popular cuisines based on the number of votes

3.Determine which cuisines consistently receive higher ratings



# Importing necessary libraries

```python
##  Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#  Data collection

```python
##  Load Dataset
df = pd.read_csv("Dataset .csv")
df.dropna(subset=['Cuisines'], inplace=True)
df.head()
```

```python
##  Analyze Cuisine vs Rating
avg_rating_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
print("Top 10 Cuisines with Highest Average Rating:")
print(avg_rating_by_cuisine)
```

```python
# Plot
plt.figure(figsize=(10,5))
avg_rating_by_cuisine.plot(kind='bar')
plt.title("Top 10 Cuisines with Highest Average Rating")
plt.xlabel("Cuisine")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##  Popular Cuisines by Vote Count
popular_cuisines = df.groupby('Cuisines')['Votes'].sum().sort_values(ascending=False).head(10)
print("\nTop 10 Popular Cuisines by Number of Votes:")
print(popular_cuisines)
```

```python
# Plot
plt.figure(figsize=(10,5))
popular_cuisines.plot(kind='bar', color='green')
plt.title("Top 10 Cuisines by Total Votes")
plt.xlabel("Cuisine")
plt.ylabel("Votes")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##  Identify Cuisines with Higher Ratings
high_rating_cuisines = df[df['Aggregate rating'] >= 4.0]['Cuisines'].value_counts().head(10)
print("\nTop 10 High Rating Cuisines (Rating >= 4.0):")
print(high_rating_cuisines)
```

```python
# Plot
plt.figure(figsize=(10,5))
high_rating_cuisines.plot(kind='bar', color='orange')
plt.title("Top 10 High Rating Cuisines (Rating â‰¥ 4.0)")
plt.xlabel("Cuisine")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# Conclusion

1.North Indian, Fast Food, and Chinese were among the most frequently occurring and voted cuisines

2.Cuisines with the highest average ratings included Italian and European, showing quality over popularity

3.The analysis highlighted the gap between popularity (votes) and satisfaction (ratings), providing useful insights for marketing and menu strategy
