# Level 2 - Task 3: Feature Engineering

# Overview of Problem Statement

To improve model performance and gain deeper insights, new features must be derived from existing data.



# Objective

1.Extract new features such as the length of restaurant name and address

2.Create binary flags for table booking and online delivery availability

3.Save a clean, feature-enhanced dataset for further analysis



# Importing necessary libraries

```python
## Import Libraries
import pandas as pd
```

#  Data collection

```python
##  Load Dataset
df = pd.read_csv("Dataset .csv")
df.dropna(subset=['Cuisines'], inplace=True)
df.head()
```

```python
## Extract Length of Restaurant Name and Address
df['Restaurant_Name_Length'] = df['Restaurant Name'].apply(len)
df['Address_Length'] = df['Address'].apply(len)
```

```python
print("Sample of new length features:")
df[['Restaurant Name', 'Restaurant_Name_Length', 'Address', 'Address_Length']].head()
```

```python
## Encode Categorical Variables into New Features
df['Has_Table_Booking_Flag'] = df['Has Table booking'].apply(lambda x: 1 if x.lower() == 'yes' else 0)
df['Has_Online_Delivery_Flag'] = df['Has Online delivery'].apply(lambda x: 1 if x.lower() == 'yes' else 0)
```

```python
print("\nSample of new encoded features:")
df[['Has Table booking', 'Has_Table_Booking_Flag', 'Has Online delivery', 'Has_Online_Delivery_Flag']].head()
```

```python
## Step 5: Save the Modified Dataset
df.to_csv("Dataset_Feature_Engineered.csv", index=False)
print("\n Feature-engineered dataset saved as 'Dataset_Feature_Engineered.csv'")
```

# Conclusion

1.Successfully created new numerical features (Restaurant_Name_Length, Address_Length)

2.Encoded categorical data into meaningful binary flags (Has_Table_Booking_Flag, Has_Online_Delivery_Flag)

3.Saved a feature-engineered dataset for use in predictive modeling and advanced analysis
