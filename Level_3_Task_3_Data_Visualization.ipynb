# Level 3 - Task 3: Data Visualization

# Overview of Problem Statement

Raw data can be difficult to interpret. Visualizations help uncover patterns, trends, and relationships between variables especially customer ratings and key restaurant features.

# Objective

1.Visualize the distribution of restaurant ratings

2.Compare average ratings across different cuisines and cities

3.Explore relationships between numerical features (e.g., votes, cost, rating) using visual tools

# Importing necessary libraries

```python
##  Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#  Data collection

```python
##  Load Dataset
df = pd.read_csv("Dataset .csv")
df.dropna(subset=['Cuisines'], inplace=True)
```

```python
##  Histogram of Aggregate Rating
plt.figure(figsize=(8, 5))
sns.histplot(df['Aggregate rating'], bins=10, kde=True, color='skyblue')
plt.title("Distribution of Aggregate Rating")
plt.xlabel("Aggregate Rating")
plt.ylabel("Count")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
##  Bar Plot of Average Ratings by Cuisine
avg_rating_by_cuisine = df.groupby('Cuisines')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
avg_rating_by_cuisine.plot(kind='bar', color='teal')
plt.title("Top 10 Cuisines by Average Rating")
plt.xlabel("Cuisines")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##  Bar Plot of Average Ratings by City
avg_rating_by_city = df.groupby('City')['Aggregate rating'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 5))
avg_rating_by_city.plot(kind='bar', color='orchid')
plt.title("Top 10 Cities by Average Rating")
plt.xlabel("City")
plt.ylabel("Average Rating")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
##  Pairplot for Relationship Exploration
sns.pairplot(df[['Aggregate rating', 'Average Cost for two', 'Votes']], kind='scatter')
plt.suptitle("Pairplot of Key Numerical Features", y=1.02)
plt.tight_layout()
plt.show()
```

# Conclusion


1.The aggregate rating distribution was slightly skewed toward mid-range values (3.0–4.0)

2.Cities like Bangalore and cuisines like Italian showed higher average ratings

3.Pairplots revealed positive correlations between votes and ratings, indicating higher-rated restaurants often receive more customer engagement

4.These visual insights can guide both business decisions and user experience optimization
